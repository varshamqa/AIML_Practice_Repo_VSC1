```python
nltk.download('wordnet')
from nltk.corpus import wordnet
```

```python
wordict = ["fair", "bat", "light", "watch", "test", "challenge","current"]
```

```python
print(wordict)
```

```python
for word in wordict:
  synonyms = []
  for syn in wordnet.synsets(word):
    for lemma in syn.lemmas():
      synonyms.append(lemma.name())
      synonyms = list(set(synonyms))
  print(set(synonyms))
```

```python
target_word = wordict[2]

synsets = wordnet.synsets(target_word)
if len(synsets) >= 3:
  third_meaning = synsets[2].definition()
  print(f"The third meaning of '{target_word}' is: {third_meaning}")
else:
  print(f"The word '{target_word}' does not have enough meanings.")
```

```python
for word in wordict:
  synonyms = []
  for syn in wordnet.synsets(word):
    for lemma in syn.lemmas():
      synonyms.append(lemma.name())
  synonyms = list(set(synonyms))
  nouns = [s for s in synonyms if wordnet.synsets(s) and wordnet.synsets(s)[0].pos() == 'n']
  print(f" Nouns for '{word}':  {nouns}")
```

```python
for words in wordict:
  synonyms = []
  verbs = []
  for syn in wordnet.synsets(words):
    for lemma in syn.lemmas():
      synonyms.append(lemma.name())
      if lemma.synset().pos() == 'v':
        verbs.append(lemma.name())
  synonyms = list(set(synonyms))
  verbs = list(set(verbs))
  # print(f"Synonyms for '{words}': {synonyms}")
  print(f"Verbs for '{wordict}': {verbs}")
```

```python
for words in wordict:
  synonyms = []
  adverbs = []
  for syn in wordnet.synsets(words):
    for lemma in syn.lemmas():
      synonyms.append(lemma.name())
      if lemma.synset().pos() == 'r':
        adverbs.append(lemma.name())
  synonyms = list(set(synonyms))
  adverbs = list(set(adverbs))
  # print(f"Synonyms for '{words}': {synonyms}")
  print(f"Adverbs for '{wordict}': {adverbs}")
```

```python
for wirds in wordict:
  synonyms = []
  adjectives = []
  for syn in wordnet.synsets(wirds):
    for lemma in syn.lemmas():
      synonyms.append(lemma.name())
      if lemma.synset().pos() == 'a':
        adjectives.append(lemma.name())
        synonyms = list(set(synonyms))
  adjectives = list(set(adjectives))
  # print(f"Synonyms for '{words}': {synonyms}")
  print(f"Adjectives for '{wordict}': {adjectives}")
```

```python
for word in wordict:
  synsets = wordnet.synsets(word)
  if synsets:
    definition = synsets[0].definition()
    print(f"The Definition of '{word}' is: {definition}")
```

```python
for word in wordict:
    synsets = wordnet.synsets(word)
    if synsets:
        hypernyms = synsets[0].hypernyms()
        hypernyms_names = [hypernym.name().split('.')[0] for hypernym in hypernyms]
        print(f"Hypernyms of {word}: {', '.join(hypernyms_names)}")
    else:
        print(f"No hypernyms found for {word}")
```

```python
for word in wordict:
    synsets = wordnet.synsets(word)
    if synsets:
        hyponyms = synsets[0].hyponyms()
        hyponyms_names = [hyponym.name().split('.')[0] for hyponym in hyponyms]
        print(f"Hyponyms of {word}: {', '.join(hyponyms_names)}")
    else:
        print(f"No hyponyms found for {word}")
```

```python
def get_hyponyms(wordict):
    synsets = wordnet.synsets(wordict)
    return [hyponym.name().split('.')[0] for hyponym in synsets[0].hyponyms()] if synsets else []

for target_word in wordict:
    hyponyms = get_hyponyms(target_word)

    if len(hyponyms) >= 2:
        hyponym1, hyponym2 = hyponyms[:2]
        similarity = wordnet.synsets(hyponym1)[0].wup_similarity(wordnet.synsets(hyponym2)[0])

        print(f"Similarity between '{hyponym1}' and '{hyponym2}' for '{target_word}': {similarity}")
    else:
        print(f"Not enough hyponyms found for {target_word}")
```
